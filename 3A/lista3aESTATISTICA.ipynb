1. Liste as diferenças entre uma lista, um dicionário, um array numpy e uma series pandas?

▪ Uma lista é uma sequência ordenada de elementos, cada qual associado a um número responsável por indicar a sua posição.
▪ Um dicionário, é uma estrutura de dados em que os elementos são pares chave:valor. A chave identifica um item e o valor armazena seu conteúdo.
▪ A diferença fundamental entre listas e dicionários é que em uma lista, os índices que determinam a posição dos elementos precisam ser inteiros, enquanto em um dicionário os índices podem ser não paenas inteiros, mas também de qualquer tipo. Além disso, em um dicionário não temos o conceito de ordem, já nas listas temos.

▪ Series, une a melhor parte de listas e dicinários. Armezena uma sequência ordenada de elementos e permite que acada elemento também seja acessado de forma simples erápida através de um rótulo(parecido com keys).
