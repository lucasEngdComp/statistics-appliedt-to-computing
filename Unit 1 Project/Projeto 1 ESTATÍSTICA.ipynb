TAREFA 3:
(a) ranking de acidentes por estado para todos os anos (um só arquivo combinando a
informação dos três anos);

```python
#primeiro, vamos somar os valores dos 3 arquivos e armazenar numa variável.
acidentes2018 = dados2018["uf"].value_counts()
acidentes2019 = dados2019["uf"].value_counts()
acidentes2020 = dados2020["uf"].value_counts()
```

```python
total = acidentes2018 + acidentes2019 + acidentes2020
total.index.name = 'uf'
total.name = 'totalAcidentes'
total.sort_values(inplace = True, ascending=True) 
```

```python
newDataframe = pd.DataFrame(total)
```

```python
newDataframe.to_csv('data.csv') # arquivo criado 
```

(b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos
considerados? (proporção = numero de ilesos ou feridos graves/ número total de
pessoas envolvidas no acidente);

```python
meses = ["January", "February", "March","April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
ilesos18ar = dados2018.groupby(pd.DatetimeIndex(dados2018.data_inversa).month)["ilesos"].sum()
ilesos18 = pd.array(ilesos18ar)
```

```python
feridos18ar = dados2018.groupby(pd.DatetimeIndex(dados2018.data_inversa).month)["feridos_graves"].sum()
feridos18 = pd.array(feridos18ar)
```

```python
ilesos19ar = dados2019.groupby(pd.DatetimeIndex(dados2019.data_inversa).month)["ilesos"].sum()
ilesos19 = pd.array(ilesos19ar)
```

```python
feridos19ar = dados2019.groupby(pd.DatetimeIndex(dados2019.data_inversa).month)["feridos_graves"].sum()
feridos19 = pd.array(feridos18ar)
```

```python
ilesos20ar = dados2020.groupby(pd.DatetimeIndex(dados2020.data_inversa).month)["ilesos"].sum()
ilesos20 = pd.array(ilesos20ar)
```

```python
feridos20ar = dados2020.groupby(pd.DatetimeIndex(dados2020.data_inversa).month)["feridos_graves"].sum()
feridos20 = pd.array(feridos20ar)
```

```python
#envolvidos
tota18ar = dados2018.groupby(pd.DatetimeIndex(dados2018.data_inversa).month)["pessoas"].sum()
total18 = pd.array(tota18ar)
```

```python
tota19ar = dados2019.groupby(pd.DatetimeIndex(dados2019.data_inversa).month)["pessoas"].sum()
total19 = pd.array(tota19ar)
```

```python
total20ar = dados2020.groupby(pd.DatetimeIndex(dados2020.data_inversa).month)["pessoas"].sum()
total20 = pd.array(total20ar)
```

```python
#propoção de ilesos nos 3 anos
proporcaoi2018 = ilesos18ar/total18
proporcaoi2019 = ilesos19ar/total19
proporcaoi2020 = ilesos20ar/total20
```

```python
#propoção de feridos nos 3 anos
proporcaof2018 = feridos18ar/total18
proporcaof2019 = feridos19ar/total19
proporcaof2020 = feridos20ar/total20
```
